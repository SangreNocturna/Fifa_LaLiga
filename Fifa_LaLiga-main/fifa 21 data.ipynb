```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
sns.set()
```

```python

def preprocess(df):
    cols = []
    for column in df.columns:
        cols.append(column.lower())
    df.columns = cols
    df.dropna(inplace=True)
    return df


df = preprocess(pd.read_csv('fifa21_train.csv'))
df.head()
```
